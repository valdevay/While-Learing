### Основы статистики

#### Генеральная совокупность и выборка

Выборка - часть генеральной совокупности (больные, которых мы можем включить в нашу выборку)   

Генеральная совокупность - все, например все больные сахарным деобетом на земле

Репрезентативность выборки - выборка, которая имеет необходимые характеристики для иследованиий

Простая случайная выборка - выборка случайных представителей из генеральной совокупности, постепенно выборка будет напоминать генеральную совокупность по своим характеристикам

Стратифицированная выборка - перед тем, как случайным образом извлекать элементы генеральной совокупности,
совокупность разбивается на несколько групп (например по полу или возрасту), а далее уже извлекаем элементы из каждой группы

Групповая выборка (clusster sample) - так же как и стратифицированная выборка делиться на группы, по по схожим характеристикам,
например районы в городе, а далее извлекаются элементы из одного района и по его характеристикам остальные районы будут схожи

+ экономия времени

#### Типы переменных

Количественные vs Качественные(Номинативные)

Количественные - измеряемые значения некого признака (например рост)
можно выполнять матемтические действия
+ количесвтвенные(непрерывные) - может принимать любое значение на каком-то промежутке (например рост может быть от 160 до 190)
это переменные, которые выражаются действительными числами.

+ количественные(дискретные) - может принимать только определенное значение (кол-во детей в семье - 1,2,3)
это переменные, которые выражаются только целыми числами.
  

Качественные
(Номитативные) - используются для разделения испытуемых или исследований на группы, все женские - 1, мужские - 0
- за цифрами не стоит математического смысла, это больше маркеры или лейблы

Ранговые переменные - когда мы понимаем, что один выше по рангу другого, но на сколько, во сколько раз вырожено его преимущество не можем сказать 
+ можем только сравнивать (дотсупно из математических операций)
